```python
#Importação das bibliotecas
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importando a base de dados
base = pd.read_csv('trees.csv')
base.shape
```

```python
#criação do histograma
h = np.histogram(base.iloc[:,1], bins = 6) #[:,1] pegando os dados da coluna 1 'height'
#bins = número de eixos, no caso 6
h
```

```python
#visualização
plt.hist(base.iloc[:,1], bins = 6)
plt.title('Arvores')
plt.ylabel('Frequencia')
plt.xlabel('Altura')
```

```python
#histograma com linha de densidade, utilizando o kde = True e a biblioteca seaborn
sns.histplot(base.iloc[:,1], kde=True, bins=6, color='red').set(title='Arvores')
```
